```python
while(True):    
    
    import math

    #taking input from user
    n1=input("enter first number:")                      #first number                    

    op=input("enter operation\(sum,div,int_div,mul,sub,mod,expo,even_odd,percentage,factorial,sqrt\) :")    #operation

    def even_odd(n1):
        "creating even odd function"
        if float(n1)%2 ==0:
            print("number is even")
        elif float(n1)%2 !=0:
            print("number is odd")
        else:
            print("invalid input")

    def factorial(n1):   # factorial function
        i=1
        fact=1
        while(i<=int(n1)):
            fact*=i
            i+=1
        print(fact)    

    def percentage(n1,n2):             #creating function to check percentage
        n=float(n1)/float(n2)
        r=n*100
        print(r)

#making conditional statements for different operations
                      
    if op =="sum":              #for sum
        n2=float(input("Enter second number:"))
        r=float(n1)+float(n2)
        print(r)

    elif op=="div":              #for divide
    
        n2=float(input("Enter second number:"))
        try:
            r=float(n1)/float(n2)
            print(r)
        except:
            print("invalid input")
    elif op=="mul":             #for multiplication
        n2=float(input("Enter second number:"))
        r=float(n1)*float(n2)
        print(r)

    elif op=="sub":             #for substraction 
        n2=float(input("Enter second number:"))
        r=float(n1)-float(n2)
        print(r)

    elif op=="expo":            #for exponent
        n2=float(input("Enter power:"))
        r=float(n1)**float(n2)
        print(r)
    
    elif op=="int_div":         #for integral divition
        n2=float(input("Enter second number:")) 
        try:
            r=float(n1)//float(n2)
            print(r)
        except:
            print("invalid input")
    elif op=="mod":             #for modulo
        try:
            r=float(n1)%float(n2)
            print(r)
        except:
            print("invalid input")
    elif op=="even_odd":        #for even odd
    
        even_odd(n1)
    
    elif op=="percentage":      #for percentage
        n2=float(input("Enter second number:"))
        percentage(n1,n2)
    
    elif op=="factorial":       #for factorial
        factorial(n1)

    elif op=="sqrt":             #for square root
        print(math.sqrt(float(n1)))
    else:
        print("invalid input")


```
